```python
a=[[1,1,1],[1,1,0],[1,0,1]]
rc=1
rr=1
color=2
```

```python
def change_form(a,rr,rc,color):
    prev=a[rr][rc]
    a[rr][rc]=color
    if rc-1>-1 and a[rr][rc-1]==prev:
        change_form(a,rr,rc-1,color)
    if rc+1<len(a[0]) and a[rr][rc+1]==prev:
        change_form(a,rr,rc-1,color)
    if rr-1>-1 and a[rr-1][rc]==prev:
        change_form(a,rr-1,rc,color)
    if rr+1<len(a) and a[rr+1][rc]==prev:
        change_form(a,rr+1,rc,color)
```

```python
change_form(a,rc,rc,color)
```
